```python
#Statistical inference
```

```python

# Null Hypothesis (H0): There is no significant difference in the mean Gross Sales between Vehicle Types.
# Alternative Hypothesis (H1): There is a significant difference in the mean Gross Sales between Vehicle Types.

# Selecting a significance level (alpha)
alpha = 0.05

# Spliting the data into two groups based on vehicle type (e.g., 'Sedan' and 'SUV')
group_1 = data[data['vechile type'] == 'Sedan']['Gross Sales']
group_2 = data[data['vechile type'] == 'SUV']['Gross Sales']

# Performing an independent two-sample t-test
t_stat, p_value = stats.ttest_ind(group_1, group_2, equal_var=False)

# Checking if the p-value is less than alpha
if p_value < alpha:
    print("Hypothesis Test Result: Reject the null hypothesis.")
    print("There is a significant difference in mean Gross Sales between Vehicle Types.")
else:
    print("Hypothesis Test Result: Fail to reject the null hypothesis.")
    print("There is no significant difference in mean Gross Sales between Vehicle Types.")

# Calculating and interpret confidence intervals for both groups
confidence_level = 0.95
mean_group_1 = group_1.mean()
mean_group_2 = group_2.mean()
std_error_group_1 = stats.sem(group_1)
std_error_group_2 = stats.sem(group_2)

conf_interval_group_1 = stats.t.interval(confidence_level, len(group_1) - 1, loc=mean_group_1, scale=std_error_group_1)
conf_interval_group_2 = stats.t.interval(confidence_level, len(group_2) - 1, loc=mean_group_2, scale=std_error_group_2)

print(f"Confidence Interval for 'Sedan' Gross Sales: ({conf_interval_group_1[0]:.2f}, {conf_interval_group_1[1]:.2f})")
print(f"Confidence Interval for 'SUV' Gross Sales: ({conf_interval_group_2[0]:.2f}, {conf_interval_group_2[1]:.2f})")
```
